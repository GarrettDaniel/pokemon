```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import json
```

```python
## Load data
pokemon = pd.read_csv('Pokemon.csv')
```

```python
pokemon.head()
```

```python
## Deal with megas since they give random types that aren't in some generations
## Like Fairy type in Generation 1 (TODO)
```

```python
## Drop unnecessary columns
pokemon.drop(columns=["#", "Name"], inplace=True)
```

```python
pokemon.describe()
```

```python
## Let's see the type distributions across all generations
for generation in pokemon['Generation'].unique():
    subset_filter = pokemon['Generation'] == generation
    pokemon_subset = pokemon[subset_filter]
    general_title = "Pokemon Type Distributions: Generation "
    
    sns.histplot(
        pokemon_subset, 
        y="Type 1", 
        binwidth=2,
    ).set(title=general_title + str(generation))
    
    plt.show()
```

```python
## Top 3 common types per Generation
most_common_types = {}
for generation in pokemon['Generation'].unique():
    subset_filter = pokemon['Generation'] == generation
    gen_subset = pokemon[subset_filter]

    type_series = gen_subset['Type 1'].value_counts()
    top_types_dict = type_series.head(3).to_dict()

    gen_string = "Generation {}".format(generation)
    most_common_types[gen_string] = top_types_dict
```

```python
print(json.dumps(most_common_types, indent=4))
```

```python
## Compare base stat totals per generation
sns.boxplot(data=pokemon, x="Generation", y="Total")
plt.show()
```

```python
## Compare all stats of legendaries vs non-legendaries each generation
pokemon.groupby(['Generation', 'Legendary']).mean()
```
